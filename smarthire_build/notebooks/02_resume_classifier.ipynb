Pipeline: cleaned resume text -> TF-IDF -> compare Logistic Regression / Random Forest / Linear SVM -> save the best model + vectorizer + label encoder to `models/`.

This notebook is a thin walkthrough around `src/models/classifier.py` — the reusable logic lives in `src/`, the notebook is for exploration, plots, and the written narrative for the report.

```python
import sys, pathlib
sys.path.append(str(pathlib.Path.cwd().parent))

import pandas as pd

from src.config import FIGURES_DIR
from src.models.classifier import (
    load_clean_resumes,
    train_and_compare,
    save_artifacts,
)
from src.evaluate import plot_confusion_matrix
```

## 1. Load cleaned resume data
(Requires `python -m src.data.preprocess` to have been run first.)

```python
df = load_clean_resumes()
print(f"{len(df)} resumes across {df['category'].nunique()} categories")
df.head()
```

## 2. Train & compare candidate models
Trains Logistic Regression, Random Forest, and Linear SVM on the same stratified train/test split and TF-IDF features, then prints per-model metrics and a classification report for each.

```python
best_name, best_model, vectorizer, label_encoder, results, split = train_and_compare(df)
best_name
```

## 3. Compare results in a table

```python
results_df = pd.DataFrame(results).T.sort_values('f1', ascending=False)
results_df
```

## 4. Confusion matrix for the best model

```python
fig = plot_confusion_matrix(
    split['y_test'],
    split['y_pred_best'],
    labels=list(range(len(label_encoder.classes_))),
    save_path=FIGURES_DIR / 'classifier_confusion_matrix.png',
)
```

## 5. Save the trained classifier + vectorizer + label encoder
These artifacts (`models/classifier.pkl`, `models/tfidf_vectorizer.pkl`, `models/label_encoder.pkl`) are what the Phase 3 Streamlit app will load for live predictions.

```python
save_artifacts(best_model, vectorizer, label_encoder)
```
